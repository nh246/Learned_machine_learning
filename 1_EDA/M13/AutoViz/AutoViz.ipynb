# Using AUtoViz for EDA

```python
import pandas as pd
import seaborn as sns
from autoviz.AutoViz_Class import AutoViz_Class
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
# 1. Instantiate the class
AV = AutoViz_Class()
```

```python
# Correctly call AutoViz using the DataFrame
df_output = AV.AutoViz(
    filename='',              # Set to empty string since we are using a DataFrame
    sep='',                   # Set to empty string (ignored if no filename)
    dfte=df,          # Pass your DataFrame here!
    depVar='survived',        # Set the target variable for the Titanic dataset
    header=0,
    verbose=1                 # Displays plots inline
)
```
